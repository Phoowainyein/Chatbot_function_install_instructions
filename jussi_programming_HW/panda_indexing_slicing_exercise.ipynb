```python
# 1 c : Use .iloc to change values in the dataframe :
import numpy as np
import pandas as pd

my_np_array=np.array([range(10),range(10,20),range(20,30)]).T


my_df =pd.DataFrame(my_np_array,columns=["first_column","second_column","third_column"])
my_df.head(10)

my_df.iloc[0,0]=997
my_df.iloc[3,0]=-997
my_df.iloc[4,0]=1000
my_df.iloc[2,1]=-999
my_df.iloc[9,1]=998
my_df.iloc[1,2]=999
my_df.iloc[6,2]=-998
my_df.iloc[8,2]=-1000
my_df

```
